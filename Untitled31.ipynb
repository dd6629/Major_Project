```python
import pandas as pd
import numpy as np
from scipy.signal import resample
def downsample_eeg(df, original_freq, new_freq):
    num_samples = len(df)
    new_num_samples = int(num_samples * new_freq / original_freq)
    
    resampled_df = pd.DataFrame()
    resampled_df['Timestamp'] = np.linspace(df['Timestamp'].min(), df['Timestamp'].max(), new_num_samples)
    
    for col in df.columns[1:]:
        resampled_df[col] = resample(df[col], new_num_samples)
    
    return resampled_df

# Downsample the data
original_freq = 256  # Original frequency
new_freq = 100    # New frequency
resampled_df = downsample_eeg(df, original_freq, new_freq)

# Print the resampled data
print(resampled_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch, welch
from scipy.fft import fft, ifft, fftfreq


def butter_bandpass(lowcut, highcut, fs, order=2):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=2):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Define the notch filter
def notch_filter(data, fs, freq=60.0, quality_factor=30.0):
    nyquist = 0.5 * fs
    freq = freq / nyquist
    b, a = iirnotch(freq, quality_factor)
    y = filtfilt(b, a, data)
    return y

# Parameters
fs = 256.0  # Sampling frequency in Hz
lowcut = 1.0  # Low cut frequency in Hz for initial bandpass filter
highcut = 75.0  # High cut frequency in Hz for initial bandpass filter

# Apply initial bandpass filter
df['TP9_filtered'] = butter_bandpass_filter(df['TP9'], lowcut, highcut, fs, order=2)

# Apply notch filter
df['TP9_filtered'] = notch_filter(df['TP9_filtered'], fs, freq=60.0, quality_factor=30.0)

# Ensure the column exists and has no NaN values
if 'TP9_filtered' in df.columns:
    df['TP9_filtered'] = df['TP9_filtered'].fillna(0)
else:
    raise KeyError("The column 'TP9_filtered' does not exist in the DataFrame.")

# Convert the filtered data to numpy array
tp9_filtered_np = df['TP9_filtered'].values

# Fourier Transform to convert the signal to the frequency domain
tp9_fft = fft(tp9_filtered_np)
frequencies = fftfreq(len(tp9_fft), 1/fs)

# Mask to isolate the theta band (4-8 Hz)
theta_mask = (frequencies >= 4) & (frequencies <= 8) | (frequencies <= -4) & (frequencies >= -8)
theta_fft = np.zeros_like(tp9_fft)
theta_fft[theta_mask] = tp9_fft[theta_mask]

# Inverse Fourier Transform to convert back to the time domain
df['TP9_theta'] = ifft(theta_fft).real

# Plot the original, filtered, and theta band TP9 signals
plt.figure(figsize=(15, 5))
plt.plot(df['Timestamp'], df['TP9'], label='Original TP9', alpha=0.5)
plt.plot(df['Timestamp'], df['TP9_filtered'], label='Filtered TP9', linewidth=2)
plt.plot(df['Timestamp'], df['TP9_theta'], label='Theta Band TP9', linewidth=2)
plt.xlabel('Timestamp')
plt.ylabel('TP9 Signal')
plt.title('Original, Filtered, and Theta Band TP9 Signal')
plt.legend()
plt.grid(True)
plt.show()

# Display the theta band values
print(df['TP9_theta'].head())
```

```python
import pandas as pd
import numpy as np
from scipy.signal import welch
from scipy.stats import entropy

# Create the dataframe as provided
df=pd.read_csv("eeg_baseline_level_1.csv")

# Define the window size in seconds
window_size = 10

# Segment the data
segments = []
start_time = df['Timestamp'].iloc[0]
end_time = df['Timestamp'].iloc[-1]

# Create windows
for start in np.arange(start_time, end_time, window_size):
    end = start + window_size
    segment = df[(df['Timestamp'] >= start) & (df['Timestamp'] < end)]
    if not segment.empty:
        segments.append(segment)

# Define functions to calculate PSD and DE
def calculate_psd(signal, fs=256):
    freqs, psd = welch(signal, fs=fs, nperseg=256)
    return psd

def calculate_de(signal):
    histogram, bin_edges = np.histogram(signal, bins=256, density=True)
    de = entropy(histogram)
    return de

# Extract features for each segment
features = []

for segment in segments:
    segment_features = {
        'start_time': segment['Timestamp'].iloc[0],
        'end_time': segment['Timestamp'].iloc[-1],
    }
    for col in ['TP9', 'AF7', 'AF8', 'TP10']:
        psd = calculate_psd(segment[col])
        de = calculate_de(segment[col])
        segment_features[f'{col}_psd'] = np.mean(psd)  # or any other summary statistic
        segment_features[f'{col}_de'] = de
    features.append(segment_features)

# Convert features to a DataFrame
features_df = pd.DataFrame(features)
print(features_df)
```
